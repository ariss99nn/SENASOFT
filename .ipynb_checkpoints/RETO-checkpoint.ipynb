```python
# Diccionario actualizado con "necesitamos más acceso a internet" en Educación
comentarios_por_categoria = {
    'Salud': [
        "faltan médicos en el centro de salud.",
        "falta agua potable en varias casas."
    ],
    'Educación': [
        "no hay suficientes escuelas públicas.",
        "no tenemos centros culturales ni bibliotecas.",
        "necesitamos más acceso a internet en la zona."
    ],
    'Seguridad': [
        "las calles están muy oscuras y peligrosas.",
        "queremos más presencia policial."
    ],
    'Medio Ambiente': [
        "las basuras no se recogen a tiempo.",
        "la contaminación del río está aumentando.",
        "hay problemas con la recolección de basura."
    ]
}

# Función de reasignación segura
def asignar_comentario(cat):
    if cat in comentarios_por_categoria:
        return np.random.choice(comentarios_por_categoria[cat])
    else:
        return np.nan

# Aplicar la reasignación usando .loc
df.loc[:, "Comentario"] = df["Categoría del problema"].apply(asignar_comentario)

# Convertir a string nativo usando .loc
df.loc[:, "Comentario"] = df["Comentario"].astype("string")
```

```python
df["Comentario"].unique()
```

```python
df = df.drop(columns=['Edad', 'ID','Género'], errors='ignore')
df['Fecha del reporte'] = pd.to_datetime(df['Fecha del reporte'], errors='coerce')
df.head()
```

```python
def extraer_por_columna(df, columna):
    """Devuelve las filas donde la columna indicada tiene valor 1."""
    return df[df[columna] == 'Educación']
```

```python
acce_df = extraer_por_columna(df, "Acceso a internet")
print(acce_df)
```
